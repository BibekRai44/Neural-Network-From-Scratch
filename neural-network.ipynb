```python
import numpy as np
input = [1,2,3,2.5]
weights = [[0.2,0.8,-0.5,1.0],[0.5,-0.91,0.26,-0.5],[-0.26,-0.27,0.17,0.87]]

bias = [2,3,0.5]

output = np.dot(weights,input) + bias

print(output)
```

```python
input4 = [[1.0,2.0,3.0,2.5],
         [2.0,5.0,-1.0,2.0],
         [-1.5,2.7,3.3,-0.8]]

weights5 = [[0.1,-0.14,0.5],
            [-0.5,-0.12,-0.33],
            [-0.44,0.73,-0.13]]
weights4 = [[0.2,0.8,-0.5,1.0],
            [0.5,-0.91,0.26,-0.5],
            [-0.26,-0.27,0.17,0.87]]
bias4 = [2,3,0.5]
bias5 = [-1,2,-0.5]

#transposing matrix

layer1_output = np.dot(input4,np.array(weights4).T ) + bias4
layer2_output = np.dot(layer1_output,np.array(weights5).T) + bias5
print(layer1_output)
print(layer2_output)
```

```python
X = [[1.0,2.0,3.0,2.5],
    [2.0,5.0,-1.0,2.0],
    [-1.5,2.7,3.3,-0.8]]

np.random.seed(0)

class Layer_Dense:
    def __init__(self,n_inputs,n_neurons):
        self.weights = 0.10 * np.random.randn(n_inputs,n_neurons)
        self.biases = np.zeros((1,n_neurons))
    def forward(self,inputs):
        self.output = np.dot(inputs,self.weights) + self.biases

layer1 = Layer_Dense(4,5)
layer2 = Layer_Dense(5,2)

layer1.forward(X)
print(layer1.output)
layer2.forward(layer1.output)
print(layer2.output)
```
